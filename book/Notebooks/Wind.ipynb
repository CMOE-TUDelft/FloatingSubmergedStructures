```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
U_10 = 25.0       # Reference mean wind speed at 10m (m/s)
z_0 = 0.0002      # Roughness length for open sea (m)
h_max = 60.0      # Maximum height to plot (m)

# --- Calculation ---
# Generate heights (z) from near surface up to max height
z = np.linspace(0.1, h_max, 500) 

# Calculate Wind Speed U(z) using the Logarithmic Law
# Formula: U(z) = U_10 * (ln(z/z0) / ln(10/z0))
U_z = U_10 * (np.log(z / z_0) / np.log(10 / z_0))

# --- Plotting ---
plt.figure(figsize=(6, 7), dpi=100) # Vertical aspect ratio matches standard profiles
plt.plot(U_z, z, color='#0077BE', linewidth=2.5, label='Logarithmic Profile')

# Add reference line at 10m
plt.axhline(y=10, color='gray', linestyle='--', alpha=0.7)
plt.scatter([U_10], [10], color='red', zorder=5)
plt.text(U_10 + 1, 10.5, f'$U_{{10}}={int(U_10)}$ m/s', color='red', fontsize=10)

# Labels and Styling
plt.title('Vertical Wind Profile (Open Sea)', fontsize=14, pad=15)
plt.xlabel('Mean Wind Speed $U_z$ [m/s]', fontsize=12)
plt.ylabel('Height Above Sea Level $z$ [m]', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(0, h_max)
plt.xlim(15, 30) # Adjust x-limits to focus on the curve area

# Save command (uncomment to save file for your markdown)
# plt.savefig('fig/wind_profile.png', bbox_inches='tight')

plt.show()
```
